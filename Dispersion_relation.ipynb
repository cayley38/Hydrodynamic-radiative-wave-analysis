# Wave number one: $p_g = 3 p_{rad}$

```python
import numpy as np

from geodesic_angular_grid import make_geodesic_angular_grid, geodesic_grid_summary
from radiative_dispersion import DispersionBackground, solve_dispersion_relation
from radiative_dispersion_directions import directional_mode_table, directional_mode_dict
from radiative_dispersion_isotropy import (
    print_background_comoving_isotropy_report,
    print_mode_comoving_isotropy_report,
)

grid = make_geodesic_angular_grid(generation=2)
print(geodesic_grid_summary(grid))

bg = DispersionBackground.from_parameters(
    rho0=1.0,
    temperature0=0.1,
    gamma_ad=5.0 / 3.0,
    absorption_opacity=10.0,
    scattering_opacity=10.0,
    erad0=0.1,
    grid=grid,
    arad=0.1/(0.1**4.0),
    require_thermal_equilibrium=True,
)

sol = solve_dispersion_relation(bg, k=2.0 * np.pi, normalize_to="rho1")
#sol.modes[0].omega
#print_mode_comoving_isotropy_report(sol.modes[30], bg, background_u1=0.0)
```

```python
#print_mode_comoving_isotropy_report(sol.modes[86], bg, background_u1=0.0)
```

```python
for i in range(len(sol.modes)):
    mode = sol.modes[i]
    if ( np.abs( np.imag(mode.omega) ) < 8 ):
        print("Wave Id = ", i, "Mode.omega = ", np.real(mode.omega), np.imag(mode.omega) )
```
